```python
from abydos.phonetic import *
from abydos.fingerprint import *
from abydos.distance import *

import pandas as pd
import numpy as np
```

```python
names = pd.read_csv('../tests/corpora/uscensus2000.csv',
                    comment='#', index_col=1, usecols=(0,1), keep_default_na=False)
names.head()
```

```python
reverse_soundex = {}
for name in names.name:
    sdx = soundex(name)
    if sdx not in reverse_soundex:
        reverse_soundex[sdx] = set()
    reverse_soundex[sdx].add(name)
```

```python
reverse_soundex[soundex('WILLIAMSON')]
```

```python
pd.DataFrame.from_dict({collision: levenshtein('WILLIAMSON', collision) for 
                        collision in reverse_soundex[soundex('WILLIAMSON')]}, 'index',
                       columns=['Levenshtein distance from WILLIAMSON'])
```
